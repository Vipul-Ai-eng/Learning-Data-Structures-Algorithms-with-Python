```python
class stack:
    def __init__(self):
        self.s = []

    def length(self):
        return len(self.s)

    def push(self,value):
        self.s.insert(0,value)

    def peek(self): #see top element 
        if len(self.s) == 0:
            raise Exception("Stack is Empty")
        else:
            return self.s[0]

    def pop(self):
        if len(self.s) == 0:
            raise Exception("Stack is Empty")
        else:
            return self.s.pop(0)
```

```python
stk = stack()
stk.pop()
```

```python
stk.push(10)
stk.push(20)
stk.push(30)
print(stk.peek())
```

```python
print(stk.pop())
print(stk.pop())
print(stk.pop())
```
